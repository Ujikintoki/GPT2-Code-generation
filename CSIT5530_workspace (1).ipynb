# **1% baseline**

```python
import os
import torch
import math
from datasets import load_from_disk
from transformers import AutoModelForCausalLM, AutoTokenizer

# ================= Configuration =================
MODEL_CHECKPOINT = "gpt2"

# 挂载 Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 设置数据路径（使用文件夹路径，不是具体的 .arrow 文件）
DATA_PATH = "/content/drive/MyDrive/GPT2-Code-generation/data/processed/debug_sample"

def main():
    # 1. Setup Device
    if torch.cuda.is_available():
        device = torch.device("cuda")
        print(f"Using GPU: {torch.cuda.get_device_name(0)}")
    else:
        device = torch.device("cpu")
        print("Using CPU. Evaluation might be slow.")

    # 2. Load the Pre-trained Model and Tokenizer
    print(f"\nLoading baseline model '{MODEL_CHECKPOINT}'...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_CHECKPOINT)
    tokenizer.pad_token = tokenizer.eos_token

    model = AutoModelForCausalLM.from_pretrained(MODEL_CHECKPOINT)
    model.to(device)
    model.eval()

    # ================= PART 1: Quantitative Evaluation (Perplexity) =================
    print(f"\n--- Part 1: Quantitative Evaluation (Perplexity) ---")
    try:
        # 检查路径是否存在
        if not os.path.exists(DATA_PATH):
            print(f"Error: Dataset not found at {DATA_PATH}")
            return

        # 加载数据集（使用文件夹路径）
        dataset = load_from_disk(DATA_PATH)
        print(f"Loaded debug dataset with {len(dataset)} chunks.")

        total_loss = 0.0
        total_batches = len(dataset)

        # 可选：限制评估数量以加快速度
        max_eval_batches = min(total_batches, 500)
        print(f"Evaluating on {max_eval_batches} chunks...")

        with torch.no_grad():
            for i in range(max_eval_batches):
                input_ids = torch.tensor([dataset[i]["input_ids"]]).to(device)
                labels = torch.tensor([dataset[i]["labels"]]).to(device)

                outputs = model(input_ids=input_ids, labels=labels)
                total_loss += outputs.loss.item()

        avg_loss = total_loss / max_eval_batches
        perplexity = math.exp(avg_loss)
        print(f"📊 Baseline Average Loss: {avg_loss:.4f}")
        print(f"📊 Baseline Perplexity (PPL): {perplexity:.4f}")

    except Exception as e:
        print(f"Error loading dataset: {e}")

    # ================= PART 2: Qualitative Evaluation =================
    print(f"\n--- Part 2: Qualitative Evaluation (Case Studies) ---")
    prompts = [
        "def calculate_average(numbers):",
        "import requests\n\ndef fetch_webpage(url):"
    ]

    for i, prompt in enumerate(prompts):
        print(f"\n[Test Case {i+1}] Prompt:\n{prompt}")

        inputs = tokenizer(prompt, return_tensors="pt").to(device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_length=60,
                num_return_sequences=1,
                pad_token_id=tokenizer.eos_token_id,
                do_sample=True,
                temperature=0.7,
                top_k=50
            )

        generated_code = tokenizer.decode(outputs[0], skip_special_tokens=True)
        print("-" * 30)
        print(generated_code)
        print("-" * 30)

if __name__ == "__main__":
    main()
```

# **100% baseline**

```python
import os
from datasets import load_dataset
from transformers import AutoTokenizer

# ================= Configuration =================
MODEL_CHECKPOINT = "gpt2"
MAX_LENGTH = 256
STRIDE = 10

# 定义输出路径
# 对于 Colab，修改为你的 Drive 路径
OUTPUT_DIR = "/content/drive/MyDrive/GPT2-Code-generation/data/processed"

def tokenize_and_chunk(examples, tokenizer):
    """
    Tokenize the raw code strings and apply a sliding window chunking strategy.
    """
    # Tokenize the entire batch without truncation
    tokenized_inputs = tokenizer(examples["whole_func_string"], add_special_tokens=False)

    input_ids_list = []
    labels_list = []

    # Apply sliding window chunking for each document
    for input_ids in tokenized_inputs["input_ids"]:
        # Only process documents longer than MAX_LENGTH
        if len(input_ids) >= MAX_LENGTH:
            # Slide the window across the document
            for i in range(0, len(input_ids) - MAX_LENGTH + 1, STRIDE):
                chunk = input_ids[i : i + MAX_LENGTH]
                input_ids_list.append(chunk)
                labels_list.append(chunk.copy())
        else:
            # 可选：处理短文档（如果需要保留所有数据）
            # 这里选择 padding 到 MAX_LENGTH
            padding_length = MAX_LENGTH - len(input_ids)
            padded_chunk = input_ids + [tokenizer.pad_token_id] * padding_length
            input_ids_list.append(padded_chunk)
            # 对于 labels，将 padding 部分设为 -100 以忽略损失计算
            labels = padded_chunk.copy()
            if padding_length > 0:
                labels[-padding_length:] = [-100] * padding_length
            labels_list.append(labels)

    return {
        "input_ids": input_ids_list,
        "labels": labels_list
    }

def main():
    print("=" * 50)
    print("Processing FULL CodeSearchNet Python dataset")
    print("=" * 50)

    # ========== 修改点 1: 加载全部训练数据 ==========
    print("\n1. Loading CodeSearchNet (Python subset) - FULL TRAIN set...")
    # 修改前: split="train[:1%]"
    # 修改后: split="train"  (加载全部)
    raw_datasets = load_dataset("code_search_net", "python", split="train")
    print(f"   Loaded {len(raw_datasets)} raw functions.")

    # ========== 修改点 2: 可选 - 创建验证集 ==========
    # 从训练集中分出 5% 作为验证集
    print("\n2. Splitting into train and validation sets (95%/5%)...")
    split_dataset = raw_datasets.train_test_split(test_size=0.05, seed=42)
    train_dataset = split_dataset["train"]
    val_dataset = split_dataset["test"]
    print(f"   Train: {len(train_dataset)} functions")
    print(f"   Validation: {len(val_dataset)} functions")

    print("\n3. Loading GPT-2 BPE Tokenizer...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_CHECKPOINT)
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.pad_token_id = tokenizer.eos_token_id

    print("\n4. Executing Tokenization and Sliding Window Chunking on TRAIN set...")
    lm_train = train_dataset.map(
        lambda examples: tokenize_and_chunk(examples, tokenizer),
        batched=True,
        remove_columns=train_dataset.column_names,
        desc="Processing train set",
    )

    print("\n5. Executing Tokenization and Sliding Window Chunking on VALIDATION set...")
    lm_val = val_dataset.map(
        lambda examples: tokenize_and_chunk(examples, tokenizer),
        batched=True,
        remove_columns=val_dataset.column_names,
        desc="Processing validation set",
    )

    print(f"\n6. Processing complete!")
    print(f"   Raw train functions: {len(train_dataset)} → Chunks: {len(lm_train)}")
    print(f"   Raw val functions: {len(val_dataset)} → Chunks: {len(lm_val)}")

    print("\n7. Saving processed data to local disk...")
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    # 保存为不同的文件名
    train_save_path = os.path.join(OUTPUT_DIR, "full_train")
    val_save_path = os.path.join(OUTPUT_DIR, "full_val")

    lm_train.save_to_disk(train_save_path)
    lm_val.save_to_disk(val_save_path)

    print(f"   Train data saved to: {train_save_path}")
    print(f"   Validation data saved to: {val_save_path}")

    # 打印统计信息
    print("\n" + "=" * 50)
    print("DATASET STATISTICS")
    print("=" * 50)
    print(f"Total training chunks: {len(lm_train):,}")
    print(f"Total validation chunks: {len(lm_val):,}")

    # 计算平均每个函数产生的 chunks
    avg_chunks_per_func_train = len(lm_train) / len(train_dataset)
    avg_chunks_per_func_val = len(lm_val) / len(val_dataset)
    print(f"Average chunks per function (train): {avg_chunks_per_func_train:.2f}")
    print(f"Average chunks per function (val): {avg_chunks_per_func_val:.2f}")

if __name__ == "__main__":
    main()
```

```python
import os
import torch
import math
from datasets import load_from_disk
from transformers import AutoModelForCausalLM, AutoTokenizer

# ================= Configuration =================
MODEL_CHECKPOINT = "gpt2"

# 挂载 Google Drive
from google.colab import drive
drive.mount('/content/drive')

# ========== 修改点: 使用完整数据路径 ==========
DATA_PATH = "/content/drive/MyDrive/GPT2-Code-generation/data/processed/full_val"  # 使用验证集
# 或者使用训练集的一部分进行评估
# DATA_PATH = "/content/drive/MyDrive/GPT2-Code-generation/data/processed/full_train"

def main():
    print("=" * 50)
    print("BASELINE EVALUATION - FULL DATASET")
    print("=" * 50)

    # 1. Setup Device
    if torch.cuda.is_available():
        device = torch.device("cuda")
        print(f"\n✅ Using GPU: {torch.cuda.get_device_name(0)}")
        print(f"   GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
    else:
        device = torch.device("cpu")
        print("\n⚠️ Using CPU. Evaluation will be very slow.")

    # 2. Load Model and Tokenizer
    print(f"\n📦 Loading baseline model '{MODEL_CHECKPOINT}'...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_CHECKPOINT)
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.pad_token_id = tokenizer.eos_token_id

    model = AutoModelForCausalLM.from_pretrained(MODEL_CHECKPOINT)
    model.to(device)
    model.eval()
    print(f"   Model loaded: {sum(p.numel() for p in model.parameters()):,} parameters")

    # ========== PART 1: Quantitative Evaluation ==========
    print(f"\n{'='*50}")
    print("PART 1: Quantitative Evaluation (Perplexity)")
    print("="*50)

    try:
        # 检查路径
        if not os.path.exists(DATA_PATH):
            print(f"❌ Error: Dataset not found at {DATA_PATH}")
            print("   Please run data_preprocess.py first.")
            return

        # 加载数据集
        print(f"📂 Loading dataset from: {DATA_PATH}")
        dataset = load_from_disk(DATA_PATH)
        total_chunks = len(dataset)
        print(f"   Loaded {total_chunks:,} chunks.")

        # ========== 修改点: 评估全部数据 ==========
        # 可以选择评估全部或部分（根据时间和资源）
        eval_mode = "full"  # 选项: "full", "sample"

        if eval_mode == "full":
            eval_chunks = total_chunks
            print(f"   Evaluating on ALL {eval_chunks:,} chunks...")
        else:
            eval_chunks = min(total_chunks, 5000)  # 采样 5000 个
            print(f"   Sampling {eval_chunks:,} chunks for evaluation...")

        total_loss = 0.0

        # 进度显示
        print("   Computing loss...")

        with torch.no_grad():
            for i in range(eval_chunks):
                # 每 1000 个打印进度
                if i % 1000 == 0 and i > 0:
                    print(f"      Progress: {i}/{eval_chunks} ({100*i/eval_chunks:.1f}%)")

                input_ids = torch.tensor([dataset[i]["input_ids"]]).to(device)
                labels = torch.tensor([dataset[i]["labels"]]).to(device)

                outputs = model(input_ids=input_ids, labels=labels)
                total_loss += outputs.loss.item()

        # 计算 Perplexity
        avg_loss = total_loss / eval_chunks
        perplexity = math.exp(avg_loss)

        print(f"\n{'='*50}")
        print("RESULTS")
        print("="*50)
        print(f"📊 Evaluation chunks: {eval_chunks:,}")
        print(f"📊 Average Loss: {avg_loss:.4f}")
        print(f"📊 Perplexity (PPL): {perplexity:.4f}")

        # 保存结果到文件
        results = {
            "model": MODEL_CHECKPOINT,
            "dataset": "CodeSearchNet Python (full)",
            "eval_chunks": eval_chunks,
            "avg_loss": avg_loss,
            "perplexity": perplexity,
            "device": str(device)
        }

        results_path = os.path.join(os.path.dirname(DATA_PATH), "baseline_results_full.txt")
        with open(results_path, "w") as f:
            f.write(str(results))
        print(f"💾 Results saved to: {results_path}")

    except Exception as e:
        print(f"❌ Error during evaluation: {e}")
        import traceback
        traceback.print_exc()

    # ========== PART 2: Qualitative Evaluation ==========
    print(f"\n{'='*50}")
    print("PART 2: Qualitative Evaluation (Case Studies)")
    print("="*50)

    prompts = [
        "def calculate_average(numbers):",
        "import requests\n\ndef fetch_webpage(url):",
        "def fibonacci(n):",
        "class Stack:\n    def __init__(self):"
    ]

    for i, prompt in enumerate(prompts):
        print(f"\n[Test Case {i+1}] Prompt:\n{prompt}")

        inputs = tokenizer(prompt, return_tensors="pt").to(device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_length=100,  # 增加生成长度
                num_return_sequences=1,
                pad_token_id=tokenizer.eos_token_id,
                do_sample=True,
                temperature=0.7,
                top_k=50
            )

        generated_code = tokenizer.decode(outputs[0], skip_special_tokens=True)
        print("-" * 50)
        print(generated_code)
        print("-" * 50)

if __name__ == "__main__":
    main()
```
